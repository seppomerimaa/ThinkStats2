```python
%matplotlib inline

import scipy.stats
import math
import random
import thinkstats2
import thinkplot
```

```python
mean1 = 1.5
sd1 = 1.0
mean2 = 6.0
sd2 = 2.0
data1 = [random.gauss(mean1, sd1) for _ in range(1000)]
data2 = [random.gauss(mean2, sd2) for _ in range(1000)]
pdf1 = thinkstats2.NormalPdf(mean1, sd1)
pdf2 = thinkstats2.NormalPdf(mean2, sd2)
```

```python
thinkplot.Pdf(pdf1, label='data 1')
thinkplot.Pdf(pdf2, label='data 2')
thinkplot.show()
```

```python
together = data1 + data2
#print together
est_pdf = thinkstats2.EstimatedPdf(together)
thinkplot.Pdf(pdf1, label='data 1')
thinkplot.Pdf(pdf2, label='data 2')
thinkplot.Pdf(est_pdf, label='together')
thinkplot.show()
```
